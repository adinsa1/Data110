```python
# import Matplot
import matplotlib.pyplot as plt

# Create List

x = [1, 2, 3, 4]
y = [ 3, 2, 12, 20]
z = [3, 4, 6, 12]
print(x)
print(y)
```

```python
# Plot (x,y)
```

```python
plt.plot(x,y, color = 'red', linestyle = '-', marker = '*', linewidth = '1')
plt.plot(x,z, color = 'blue', linestyle = '-', linewidth = '2', marker = '8' )
plt.legend()
plt.xlabel('Time')
plt.ylabel('Speed')
plt.title('MY First Graph')
plt.grid()
plt.show()
```
