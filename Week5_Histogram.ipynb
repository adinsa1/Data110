```python
# import packages
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/adinsa1/Data110-32213/main/dataset/titanic.csv')
# The first 5 lines
df.head()
```

```python
# The last 5 lines
df.tail()
```

```python
#find out mean, count etc
df.describe()
```

```python
# Number of Columns
df.info()
```

```python
# Draw Histogram (Age)
plt.hist(df['age'])
plt.show()
```

```python
# Show graph with length of interval, bins = 1
plt.hist(df['age'], bins=1)
plt.show()
```

```python
# Show graph wit bins = 20 and edgecolor = black
plt.hist(df['age'], bins=20, edgecolor='black')
plt.show()
```

```python
# Edgecolor , white
plt.hist(df['age'], bins=20, edgecolor='white')
plt.figure(figsize=(10,6))
plt.show()
```

```python
# use ggplot
plt.hist(df['age'], edgecolor='white')
plt.figure(figsize=(10,6))
plt.style.use('ggplot')
plt.show()
```

```python
# add new xticks
plt.hist(df['age'])
plt.xticks([0,20,40,60])
plt.show()
```

```python
# The entire graph
plt.style.use('ggplot')
plt.figure(figsize = (10, 6))
plt.hist(df['age'], bins=20, color='skyblue', edgecolor='white')
plt.title('Age Distribution of Titanic Passengers')
plt.xlabel('Age in Year')
plt.ylabel('Count')
plt.grid(axis='x')
plt.xticks([0,20,40,60])
plt.tick_params(axis='x', colors = 'gray')
plt.show

```
