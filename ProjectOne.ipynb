```python
# Import Libraries:
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# Load Dataset - Household Debt
df=pd.read_csv('https://raw.githubusercontent.com/adinsa1/Data110/main/Dataset/household_debt.csv')

# Summary of the First Five Lines
df.head()
```

```python
# Summary of the Last Five Lines
df.tail()
```

```python
# The Number and Type of Columns
df.info()
```

```python
# Statistical Characteristics of the Dataset
df.describe()
```

```python
# Plot Bar Graph of  Mortgage Debt by Period
plt.figure(figsize=(10, 15))
sorted_df = df.sort_values(by='Period',ascending=False)
plt.barh(sorted_df['Period'],sorted_df['Mortgage'], color = 'yellow')
plt.title('Mortgage Debt by Period')
plt.xlabel('Amount')
plt.ylabel('Year by Quarter')
plt.show()
```

```python

x = np.arange(5)
y1 = [0.24, 0.24, 0.25, 0.25, 0.26]
y2 = [1.38, 1.41, 1.41, 1.44, 1.46]
width = 0.5

# plot data in grouped manner of bar type
plt.figure(figsize=(10, 6))
plt.bar(x-0.2, y1, width, color='green')
plt.bar(x, y2, width, color='red')
plt.title('Comparison of the First and Last Five Lines')
plt.xticks(x, ['Q1', 'Q2', 'Q3', 'Q4', 'Q1'])
plt.xlabel("Student Loan")
plt.ylabel("Amount")
plt.legend(["First Five Lines", "Last Five Lines"])
plt.show()
```
