```python
# Import Packages
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# load dataset
df=pd.read_csv('https://raw.githubusercontent.com/adinsa1/Data110/main/Dataset/household_debt.csv')
df.head()
```

```python
plt.figure(figsize=(20, 10))
plt.bar(df['Period'],df['Mortgage'])
```

```python
plt.figure(figsize=(10, 15))

plt.barh(df['Period'],df['Mortgage'])
```

```python
plt.figure(figsize=(10, 15))

sorted_df = df.sort_values(by='Period',ascending=False)

plt.barh(sorted_df['Period'],sorted_df['Mortgage'])
```
