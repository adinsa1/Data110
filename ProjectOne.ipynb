```python
# Import Libraries:
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# Load Dataset - Household Debt
df=pd.read_csv('https://raw.githubusercontent.com/adinsa1/Data110/main/Dataset/household_debt.csv')

# Summary of the First Five Lines
df.head()
```

```python
# Summary of the Last Five Lines
df.tail()
```

```python
# Statistical Characteristics of the Dataset
df.describe()
```

```python
# Line Graph of Total Household Debt by Period
plt.figure(figsize=(10, 15))
sorted_df = df.sort_values(by='Period',ascending=False)
plt.barh(sorted_df['Period'],sorted_df['Total'], color = 'green')
plt.title('Total Household Debt by Period')
plt.xlabel('Amount')
plt.ylabel('Year by Quarter')
ax=plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.show()
```
