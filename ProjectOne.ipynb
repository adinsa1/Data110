```python
# Import Libraries:
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
# Load Dataset - Household Debt
df=pd.read_csv('https://raw.githubusercontent.com/adinsa1/Data110/main/Dataset/household_debt.csv')
# Summary of the First Five Lines
df.head()
```

```python
# Summary of the Last Five Lines
df.tail()
```

```python
# The Number and Type of Columns
df.info()
```

```python
# Statistical Characteristics of the Dataset
df.describe()
```

```python
# Line Graph of Total Household Debt by Period
plt.figure(figsize=(10, 15))
sorted_df = df.sort_values(by='Period',ascending=False)
plt.barh(sorted_df['Period'],sorted_df['Total'], color = 'green')
plt.title('TOTAL HOUSEHOLD DEBT BY YEAR', color = 'red')
plt.xlabel('Amount', color = 'red')
plt.ylabel('Year by Quarter', color = 'red')
ax=plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
plt.show()
```

```python
# Plot Bar Graph of  Mortgage Debt by Period
plt.figure(figsize=(10, 15))
sorted_df = df.sort_values(by='Period',ascending=False)
plt.barh(sorted_df['Period'],sorted_df['Mortgage'], color = 'yellow')
plt.title('MORTGAGE DEBT BY YEAR', color = 'red')
plt.xlabel('Amount', color = 'red')
plt.ylabel('Year by Quarter', color = 'red')
ax=plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
plt.show()
```

```python

# Plot Bar Graph of HE Revolving by Period
plt.figure(figsize=(10, 15))
sorted_df = df.sort_values(by='Period',ascending=False)
plt.barh(sorted_df['Period'],sorted_df['HE Revolving'], color = 'pink')
plt.title('HE REVOLVING DEBT BY YEAR', color = 'red')
plt.xlabel('Amount', color = 'red')
plt.ylabel('Year by Quarter', color = 'red')
ax=plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
plt.show()

```

```python

# Plot Bar Graph of Period and Auto Loan
plt.figure(figsize=(10, 15))
sorted_df = df.sort_values(by='Period',ascending=False)
plt.barh(sorted_df['Period'],sorted_df['Auto Loan'], color = 'skyblue')
plt.title('AUTO LOAN BY YEAR', color = 'red')
plt.xlabel('Amount', color = 'red')
plt.ylabel('Year by Quarter', color = 'red')
ax=plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
plt.show()
```
