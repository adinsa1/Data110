```python
# Part 1: Data Exploration and Visualization
# 1.1. Loading Data
# 1.1.1. Load Library
# 1.1.2. Import Iris dataset

import  seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = sns.load_dataset('iris')
```

```python
# 1.1.3. View the First Few Lines of the Dataset
df.head()
```

```python
# 1.1.4. View the Last Five Lines of the DataFrame
df.tail()
```

```python
# 1.2. Data Summaries
df.describe()
```

```python
# 1.3. Visualizations:
# 1.3.1. Create Histogram for each Feature:

# 1.3.1.1. Plotting histogram - Sepal Length
plt.style.use ('ggplot')
plt.figure(figsize = (10,8))
plt.hist(df['sepal_length'], bins=15, color='skyblue',edgecolor='white')
plt.title ('Sepal Length Distribution of the Iris Species')
plt.xlabel('Length')
plt.ylabel('Count')
plt.show()
```

```python
# 1.3.1.2. Plotting Histogram - Sepal Width

plt.style.use ('ggplot')
plt.figure(figsize = (10,8))
plt.hist(df['sepal_width'], bins=15, color='red',edgecolor='white')
plt.title ('Sepal Width Distribution of the Iris Species')
plt.xlabel('Widthth')
plt.ylabel('Count')
plt.show()

```

```python
# 1.3.1.3 Plotting Histogram - Petal Length

plt.style.use ('ggplot')
plt.figure(figsize = (10,8))
plt.hist(df['petal_length'], bins=15, color='green',edgecolor='white')
plt.title ('Petal Length Distribution of the Iris Species')
plt.xlabel('Length')
plt.ylabel('Count')
plt.show()
```

```python
# 1.3.1.4. Plotting Histogram - Petal Width

plt.style.use ('ggplot')
plt.figure(figsize = (10,8))
plt.hist(df['petal_width'], bins=15, color='yellow',edgecolor='white')
plt.title ('Petal Length Distribution of the Iris Species')
plt.xlabel('Length')
plt.ylabel('Count')
plt.show()
```

```python
# 1.3.2 Box Plot for Each Feature
# 1.3.2.1 Box Plot - Sepal Length and Sepal Width

sns.boxplot(x='sepal_length', y='sepal_width', data=df, color="skyblue")
plt.xticks(rotation=90)
plt.title('Sepal Length and  Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

plt.tight_layout()
plt.show()
```

```python
# 1.3.2.2. Sepal Length and Petal Length

sns.boxplot(x='sepal_length', y='petal_length', data=df, color="blue")
plt.xticks(rotation=90)
plt.title('Sepal Length and  Petal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')
plt.tight_layout()
plt.show()
```

```python
# 1.3.2.3. Sepal Length and Petal Width

sns.boxplot(x='sepal_length', y='sepal_width', data=df, color="r")
plt.xticks(rotation=90)
plt.title('Sepal Length and  Petal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')
plt.tight_layout()
plt.show()
```

```python
# 1.3.2.4. Sepal Width and Petal Width

sns.boxplot(x='sepal_width', y='petal_width', data=df, color='pink')
plt.xticks(rotation=90)
plt.title('Sepal Width and  Petal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
plt.tight_layout()
plt.show()
```

```python
# 1.3.2.5. Petal Length and Petal Width

sns.boxplot(x='petal_length', y='petal_width', data=df, color='green')
plt.xticks(rotation=90)
plt.title('Petal Length and  Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.tight_layout()
plt.show()

```

```python
# 1.3.2.6. Sepal Width and petal Length

sns.boxplot(x='sepal_width', y='petal_length', data=df, color='yellow')
plt.xticks(rotation=90)
plt.title('Sepal Width and  Petal Length')
plt.xlabel('Sepal Width')
plt.ylabel('Petal Length')
plt.tight_layout()
plt.show()
```

```python
# 2. Detailed Analysis by Species
# 2.1. Comparative Analysis
# 2.1.1. Scatter Plot - Sepal Length and Sepal Width

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', color='blue')
plt.title('Scatter Plot -  Between Sepal Length and Width')
plt.grid(True)
plt.show()
```

```python
# 2.1.2. Scatter Plot - Petal Length and  Width

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='petal_length', y='petal_width', color='red')
plt.title('Scatter Plot -  Between Petal Length and Width')
plt.grid(True)
plt.show()
```

```python
# 2.2. Correlation Analysis
# 2.2.1. Correlation Between Sepal Length and Sepal Width

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='sepal_length', y='sepal_width', color='blue')
sns.regplot(data=df, x='sepal_length', y='sepal_width', scatter=False, color='orange')
plt.title('Correlation Between Sepal Length and Width')
plt.grid(True)
plt.show()
```

```python
#2.2. Correlation Between Petal Length and Width

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='petal_length', y='petal_width', color='green')
sns.regplot(data=df, x='petal_length', y='petal_width', scatter=False, color='red')
plt.title('Correlation Between Petal Length and Width')
plt.grid(True)
plt.show()
```
