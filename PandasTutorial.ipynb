```python
## How to create DataFrame:
# first create nested list
students_list = [['Bob', 'Jake','Computer Engineering', 96],
               ['Ema', 'John', 'Math',78],
               ['Sara', 'Emy', 'Economics']]
```

```python
students_df = pd.DataFrame(students_list)
students_df
```

```python
# How to specify Columns
students_df = pd.DataFrame(students_list, columns= ['First_Name', 'Last_Name', 'Major', 'Score'])
students_df
```

```python
# to change index to One,Two,Three
students_df = pd.DataFrame(students_list, columns= ['First_Name', 'Last_Name', 'Major', 'Score'],
                           index = ["One", "Two", "Three"])

students_df
```

```python
# Create DataFrame, above, using Dictionary instead:
students_dict = {'First_Name': ['Bob', 'Ema', 'Sara'],
               'Last_Name': ['Jake','John', 'Emy'],
               'Major': ['Computer Engineering', 'Math', 'Economics'],
               'Score': ['96', '78', '0'],}

students_df2 = pd.DataFrame(students_dict)
students_df2
```

```python
# to retrieve Last Name and Score only:
students_df2 = pd.DataFrame(students_dict, columns=['Last_Name', 'Score'])
students_df2
```
